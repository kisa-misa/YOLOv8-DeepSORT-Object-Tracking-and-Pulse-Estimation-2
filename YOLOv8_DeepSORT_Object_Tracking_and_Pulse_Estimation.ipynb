# Make Display the Time Series Video (this section is currently not working)

```python
import cv2

image_folder = '/content/time_series'
figures_list = os.listdir(image_folder)

#grab last 2 characters of the file name:
def last_10chars(x):
    return(x[7:17])

srtd = sorted(figures_list, key = last_10chars)
video_name = 'time_series.avi'

images = [img for img in srtd if img.endswith(".png")]
frame = cv2.imread(os.path.join(image_folder, images[0]))
height, width, layers = frame.shape
fourcc = cv2.VideoWriter_fourcc(*'XVID')

video = cv2.VideoWriter(video_name, fourcc, 24, (width,height))

for image in images:
    src = image
    image = cv2.imread(os.path.join(image_folder, image))
    image = cv2.resize(image, (1280,720))
    video.write(cv2.imread(os.path.join(image_folder, src)))

cv2.destroyAllWindows()
video.release()
```

```python
from IPython.display import HTML
from base64 import b64encode
import os

# Input video path
save_path = '/content/time_series.avi'

# Compressed video path
compressed_path = "/content/result_compressed.mp4"

os.system(f"ffmpeg -i {save_path} -vcodec libx264 {compressed_path}")

# Show video
mp4 = open(compressed_path,'rb').read()
data_url = "data:video/mp4;base64," + b64encode(mp4).decode()
HTML("""
<video width=400 controls>
      <source src="%s" type="video/mp4">
</video>
""" % data_url)
```
